Correlation and P-value Between Vehicle Features and MSRP

```python
for param in ['Engine HP','Engine Cylinders','Number of Doors','highway MPG','city mpg','Popularity']:
    pearson_coef, p_value = stats.pearsonr(df[param], df['MSRP'])
    print(param)
    print("The Pearson Correlation Coefficient for ",param," is", pearson_coef, " with a P-value of P =", p_value)
```
